```python
df_pagewise = df[df.company_name == 'Sharma%20Shubham'].groupby('page_name').agg({
    'user_id_id':'count', 
    'time':'mean'
})
```

```python
df_stallwise = df[(df.company_name == 'Sharma%20Shubham') & (df.page_name.str.contains('stall_'))].groupby(['page_name','stall_name']).agg({
    'user_id_id':'count', 
    'time':'mean'
})
```
